```python
import pandas as pd

# Read the CSV file
df = pd.read_csv("archive\emails.csv", encoding="utf-8")

# List all the fields (columns)
fields = df.columns.tolist()

# Print the fields
print("Fields in the emails.csv file:")
for field in fields:
    print(field)
```

```python
# Extract unique sender's email addresses from the MetadataFrom column
unique_senders = df["ExtractedFrom"].unique()

# Print the unique sender's email addresses
print("Unique sender's email addresses:")
for sender in unique_senders:
    print(sender)
```
